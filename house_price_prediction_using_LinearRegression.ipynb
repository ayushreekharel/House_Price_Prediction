```python
X=np.array(X)   #features
Y=np.array(Y)
```

```python
def compute_cost(x,y,w,b):
  m=x.shape[0]
  cost=0
  for i in range(m):
    f_wb=w*x[i]+b
    cost=cost+(f_wb-y[i])**2
  total_cost=1/(2*m) * cost
  return total_cost
```

```python
def compute_gradient(x,y,w,b):
  m=x.shape[0]
  dj_dw=0
  dj_db=0
  for i in range(m):
    f_wb=w*x[i]+b
    dj_dw_i=(f_wb-y[i])*x[i]
    dj_db_i=f_wb-y[i]
    dj_db+=dj_db_i
    dj_dw+=dj_db_i
  dj_dw=dj_dw/m
  dj_db=dj_db/m

  return dj_dw,dj_db
```

```python
def gradient_descent(x,y,w_in,b_in,alpha,num_iters,compute_cost,compute_gradient):

  b=b_in
  w=w_in



# Loop for gradient descent
  for i in range(num_iters):
    # Compute gradients
    dj_dw, dj_db = compute_gradient(x,y,w,b)

    # Update parameters using gradients
    b = b - alpha * dj_db
    w = w - alpha * dj_dw


    return w, b



```

```python
#initialize parameters

w_init=0
b_init=0
iterations=10000
tmp_alpha=1.0e-2

w_final,b_final=gradient_descent(X,Y,w_init,b_init,tmp_alpha,iterations,compute_cost,compute_gradient)
print("w_final is:",w_final)
print("b_final is:",b_final)
```
